Task 1 :- User Overview Analysis

Task 1.1 :- Overview of the users behavior

```python
# Aggregating the required information per user

# Grouping data by 'MSISDN/Number' which represents each user
user_group = df.groupby('MSISDN/Number')

# Aggregating the required information
aggregated_data = user_group.agg(
    number_of_xDR_sessions = pd.NamedAgg(column='Bearer Id', aggfunc='count'),
    total_session_duration = pd.NamedAgg(column='Dur. (ms)', aggfunc='sum'),
    total_DL_data = pd.NamedAgg(column='Total DL (Bytes)', aggfunc='sum'),
    total_UL_data = pd.NamedAgg(column='Total UL (Bytes)', aggfunc='sum'),
    social_media_DL = pd.NamedAgg(column='Social Media DL (Bytes)', aggfunc='sum'),
    social_media_UL = pd.NamedAgg(column='Social Media UL (Bytes)', aggfunc='sum'),
    google_DL = pd.NamedAgg(column='Google DL (Bytes)', aggfunc='sum'),
    google_UL = pd.NamedAgg(column='Google UL (Bytes)', aggfunc='sum'),
    email_DL = pd.NamedAgg(column='Email DL (Bytes)', aggfunc='sum'),
    email_UL = pd.NamedAgg(column='Email UL (Bytes)', aggfunc='sum'),
    youtube_DL = pd.NamedAgg(column='Youtube DL (Bytes)', aggfunc='sum'),
    youtube_UL = pd.NamedAgg(column='Youtube UL (Bytes)', aggfunc='sum'),
    netflix_DL = pd.NamedAgg(column='Netflix DL (Bytes)', aggfunc='sum'),
    netflix_UL = pd.NamedAgg(column='Netflix UL (Bytes)', aggfunc='sum'),
    gaming_DL = pd.NamedAgg(column='Gaming DL (Bytes)', aggfunc='sum'),
    gaming_UL = pd.NamedAgg(column='Gaming UL (Bytes)', aggfunc='sum'),
    other_DL = pd.NamedAgg(column='Other DL (Bytes)', aggfunc='sum'),
    other_UL = pd.NamedAgg(column='Other UL (Bytes)', aggfunc='sum')
)

# Calculating the total data volume for each application
aggregated_data['total_social_media_data'] = aggregated_data['social_media_DL'] + aggregated_data['social_media_UL']
aggregated_data['total_google_data'] = aggregated_data['google_DL'] + aggregated_data['google_UL']
aggregated_data['total_email_data'] = aggregated_data['email_DL'] + aggregated_data['email_UL']
aggregated_data['total_youtube_data'] = aggregated_data['youtube_DL'] + aggregated_data['youtube_UL']
aggregated_data['total_netflix_data'] = aggregated_data['netflix_DL'] + aggregated_data['netflix_UL']
aggregated_data['total_gaming_data'] = aggregated_data['gaming_DL'] + aggregated_data['gaming_UL']
aggregated_data['total_other_data'] = aggregated_data['other_DL'] + aggregated_data['other_UL']

# Dropping intermediate columns
aggregated_data.drop(columns=['social_media_DL', 'social_media_UL', 
                              'google_DL', 'google_UL', 'email_DL', 
                              'email_UL', 'youtube_DL',  'youtube_UL',
                              'netflix_DL', 'netflix_UL', 'gaming_DL', 'gaming_UL',
                              'other_DL', 'other_UL'], inplace=True)

aggregated_data.head()
    
```

```python
# Identify Missing Values

null_values = df.isnull().sum()

null_values
```

```python
# Percentage % of missing values

columns_with_na = (df.isnull().sum() / len(df)) * 100

# Filter out the features and sort

columns_with_na = columns_with_na[columns_with_na > 0].sort_values(ascending=False)

# Print the feature name and the percentage of missing values

print(columns_with_na)
```

As I can see there are many features which have more than 50 % missing values. We can consider dropping these values.

```python
# Dropping the columns which have more than 50% missing values

columns_to_drop  = df.columns[df.isnull().mean() > 0.5]

data_cleaned = df.drop(columns=columns_to_drop)

# Displaying the information of the cleaned Dataframe
data_cleaned.info()
```

```python
# Replacing missing values with the mean of corresponding column

# Calculate the mean only for numerical columns
numerical_means = data_cleaned.select_dtypes(include=['float64', 'int64']).mean()

# Fill missing values in numerical columns with their respective means
df_filled = data_cleaned.fillna(numerical_means)

df_filled.isnull().sum()
```

```python
# Treating missing values of categorical data with the Mode of corresponding column

# Replace missing values in the column 'Last Location Name' with its mode
mode_of_lastlocation = df_filled['Last Location Name'].mode()[0]
df_filled['Last Location Name'].fillna(mode_of_lastlocation, inplace=True)

# Replace missing values in the column 'Last Location Name' with its mode
mode_of_handset_man = df_filled['Handset Manufacturer'].mode()[0]
df_filled['Handset Manufacturer'].fillna(mode_of_handset_man, inplace=True)

# Replace missing values in the column 'Handset Type' with its mode
mode_of_handset_type = df_filled['Handset Type'].mode()[0]
df_filled['Handset Type'].fillna(mode_of_handset_type, inplace=True)

df_filled.isnull().sum()
```

The missing values successfull identify and replaced. 

```python
# Outlier Detection and Treatment

# Function to detect and treat outliers
def replace_outliers_with_mean(dataframe):
    # Select only numeric columns (exclude datetime and non-numeric columns)
    numeric_cols = dataframe.select_dtypes(include=[np.number]).columns
    numeric_df = dataframe[numeric_cols].copy()

    # Calculate IQR for numeric columns
    Q1 = numeric_df.quantile(0.25)
    Q3 = numeric_df.quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers
    is_outlier = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

    # Replace outliers with NaN
    numeric_df[is_outlier] = np.nan

    # Replace NaN with column mean
    for column in numeric_df.columns:
        numeric_df[column].fillna(numeric_df[column].mean(), inplace=True)
    
    # Update the original dataframe
    dataframe[numeric_cols] = numeric_df

    return dataframe

# Treating outliers in the dataset
data_treated = replace_outliers_with_mean(df_filled)

# Displaying the summary statistics of the treated dataset to verify the treatment
data_treated.describe()
```
